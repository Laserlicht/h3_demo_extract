```python
import urllib.request
import tempfile
import os
import shutil
import pathlib

from lodextract.lodextract import unpack_lod
from lodextract.defextract import extract_def
```

```python
macdemo = urllib.request.urlopen("https://archive.org/download/heroes_3_complete_demo/heroes_3_complete_demo.sit").read()
```

```python
windemo = urllib.request.urlopen("https://archive.org/download/HeroesOfMightAndMagicIiiTheRestorationOfErathia/h3demo.exe").read()
```

```python
td = tempfile.TemporaryDirectory()
```

```python
os.chdir(td.name)
td.name
#td.cleanup()
```

```python
open("h3demo_mac.sit", 'wb').write(macdemo)
open("h3demo_win.exe", 'wb').write(windemo)
```

```python
os.system("unar " + "h3demo_mac.sit")
os.system("unar " + "h3demo_win.exe")
os.system("unshield x Disk1/data1.cab")
```

```python
if not os.path.exists("_h3-files"):
    os.mkdir("_h3-files")

if not os.path.exists("_h3-files/mac"):
    os.mkdir("_h3-files/mac")

if not os.path.exists("_h3-files/win"):
    os.mkdir("_h3-files/win")

if not os.path.exists("_h3-files/mac/music"):
    shutil.copytree("Heroes III Demo/music", "_h3-files/mac/music")

if not os.path.exists("_h3-files/win/music"):
    shutil.copytree("Program_Files/MP3", "_h3-files/win/music")

if not os.path.exists("_h3-files/mac/maps"):
    shutil.copytree("Heroes III Demo/maps", "_h3-files/mac/maps")

if not os.path.exists("_h3-files/win/maps"):
    shutil.copytree("Program_Files/Maps", "_h3-files/win/maps")

if not os.path.exists("_h3-files/mac/data"):
    shutil.copytree("Heroes III Demo/data", "_h3-files/mac/data")

if not os.path.exists("_h3-files/win/data"):
    shutil.copytree("Program_Files/Data", "_h3-files/win/data")

for f in pathlib.Path("_h3-files/mac").rglob("*.rsrc") :
    f.unlink()
```

```python
for path in ["_h3-files/mac/data", "_h3-files/win/data"]:
    for filename in os.listdir(path):
        if not os.path.exists(os.path.join(path, "extract", filename)):
            os.makedirs(os.path.join(path, "extract", filename), exist_ok=True)
        unpack_lod(os.path.join(path, filename), os.path.join(path, "extract", filename))
```

```python
for path in ["_h3-files/mac/data/extract/h3sprite.lod", "_h3-files/mac/data/extract/h3ab_spr.lod", "_h3-files/win/data/extract/H3SPRITE.LOD"]:
    if not os.path.exists(path + "_extracted"):
        os.makedirs(path + "_extracted", exist_ok=True)
    for filename in os.listdir(path):
        if filename.endswith(".def"):
            extract_def(os.path.join(path, filename), path + "_extracted")
```

```python
for path in ["_h3-files/win/data/extract/H3BITMAP.LOD", "_h3-files/mac/data/extract/h3bitmap.lod"]:
    shutil.copytree(path, "_h3_res/data", dirs_exist_ok=True)
for path in ["_h3-files/win/data/extract/HEROES3.SND", "_h3-files/mac/data/extract/heroes3.snd"]:
    shutil.copytree(path, "_h3_res/sound", dirs_exist_ok=True)
for path in ["_h3-files/win/data/extract/VIDEO.VID", "_h3-files/mac/data/extract/video.vid"]:
    shutil.copytree(path, "_h3_res/video", dirs_exist_ok=True)
for path in ["_h3-files/win/data/extract/H3SPRITE.LOD_extracted", "_h3-files/mac/data/extract/h3ab_spr.lod_extracted", "_h3-files/mac/data/extract/h3sprite.lod_extracted"]:
    shutil.copytree(path, "_h3_res/sprites", dirs_exist_ok=True)
```
